```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
from PIL import Image
import cv2
```

```python
# load the image
img = load_img('img .jpg')
# convert to numpy array
data = img_to_array(img)
# expand dimension to one sample
samples = expand_dims(data, 0)
# create image data augmentation generator
```

```python
def horizontal_flip(image, data, samples):
    datagen = ImageDataGenerator(horizontal_flip=True)
    it = datagen.flow(samples, batch_size=1)
    for i in range(2):
        pyplot.subplot(330 + 1 + i)
        batch = it.next()
        image = batch[0].astype('uint8')
        pyplot.imshow(image)
    pyplot.show()
    
    
horizontal_flip(img, data, samples)
```

```python
def vertical_flip(image, data, samples):
    datagen = ImageDataGenerator(vertical_flip=True)
    it = datagen.flow(samples, batch_size=1)
    for i in range(2):
        pyplot.subplot(330 + 1 + i)
        batch = it.next()
        image = batch[0].astype('uint8')
        pyplot.imshow(image)
    pyplot.show()
    
    
vertical_flip(img, data, samples)
```

```python
def roatate(image, data, samples):
    datagen = ImageDataGenerator(rotation_range=90)
    it = datagen.flow(samples, batch_size=9)
    for i in range(9):
        
        pyplot.subplot(330 + 1 + i)
        batch = it.next()
        image = batch[0].astype('uint8')
        pyplot.imshow(image)
        
    pyplot.show()


roatate(img,data,samples)
```

```python
def brightness_range(image, data, samples):
    datagen = ImageDataGenerator(brightness_range=[0.2,1.5])
    it = datagen.flow(samples, batch_size=9)
    for i in range(9):
        pyplot.subplot(330 + 1 + i)
        batch = it.next()
        image = batch[0].astype('uint8')
        pyplot.imshow(image)
    pyplot.show()
    

brightness_range(img,data,samples)
```

```python
def zoom(image, data, samples):
    datagen = ImageDataGenerator(zoom_range=[0.5,1.0])
    it = datagen.flow(samples, batch_size=9, save_to_dir='images', save_prefix='aug', save_format='png')
    for i in range(9):
        pyplot.subplot(330 + 1 + i)
        batch = it.next()
        image = batch[0].astype('uint8')
        pyplot.imshow(image)
    # show the figure
    pyplot.show()
    
    
zoom(img, data, samples)
```
